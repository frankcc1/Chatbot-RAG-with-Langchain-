ESTRUCTURA DE DOCUMENTOS

```python
#Leyendo un directorio completo
#Se puede leer desde cuallquier lado, S3 , blogstorage (Revisar documentacion)
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader

dir_loader = DirectoryLoader(
    "../data/pdf",
    glob = "**/*.pdf", #Leer archivos pdf.Para leer distintos archivos puedes crear una lista
    loader_cls = PyPDFLoader,#Aqui tambien una lista
    show_progress = False
)

documents = dir_loader.load()
documents
```
